# Titration Curves

### Importing libraries to draw & collect data

```python
import matplotlib.pyplot as plt
import pandas as pd


CB91_Green = '#47DBCD'
CB91_Pink = '#F3A0F2'

%matplotlib inline
```

### Sorting Data into two sets

```python
x = pd.read_csv("sawb.csv", usecols = [1])
y = pd.read_csv("sawb.csv", usecols = [0])
x1 = pd.read_csv("wasb.csv", usecols = [1])
y1 = pd.read_csv("wasb.csv", usecols = [0])
```

## Strong Acid - Weak Base

```python
plt.plot(x, y, 'm,')
plt.scatter(x, y, s = 200, alpha = .1, color = CB91_Pink)
plt.grid()
```

## Weak Acid - Strong Base

```python
plt.plot(x1, y1, 'b,')
plt.scatter(x1, y1, s = 200, alpha = .1, color = CB91_Green)
plt.grid()
```
